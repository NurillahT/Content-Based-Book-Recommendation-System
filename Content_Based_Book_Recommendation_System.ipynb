## **Modeling**

Menggunakan TF-IDF untuk mengubah kalimat Genre menjadi vektor.

```python
# import TF-IDF dan vektorisasi
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer()
tfidf_matrix = tf.fit_transform(clean_books['Genre'])
```

```python
print("n_samples: %d, n_features: %d" % tfidf_matrix.shape)
```

```python
# menampilkan hasil TF-IDF
feature_names = tf.get_feature_names_out()
column_names = [n for n in clean_books['Genre']]
tfidf_result = pd.DataFrame(tfidf_matrix.T.todense(), index=feature_names, columns=column_names)
tfidf_result.head(20)
```

```python
print(tfidf_matrix)
```

Menghitung cosine similarity antar buku untuk menemukan buku dengan tingkat kesamaan genre yang tinggi.

```python
# import cosine similarity dan menghitung nilai cosine similarity antar buku
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarities = cosine_similarity(tfidf_matrix)
cosine_similarities
```

```python
# menampilkan hasil perthitungan cosine similarity
cosine_sim_df = pd.DataFrame(cosine_similarities, index=clean_books['Title'], columns=clean_books['Title'])
cosine_sim_df.sample(6, axis=1).sample(10, axis=0)
```

## **Recommendation System**

Membuat fungsi untuk mendapatkan hasil rekomendasi berdasarkan cosine similarity.

```python
def recommend_books(book_title, cosine_sim=cosine_similarities):
    # mendapatkan index judul buku yang sesuai dengan judul yang dicari
    idx = clean_books.index[clean_books['Title'] == book_title].tolist()[0]

    # mengumpulkan nilai cosine similarity pada buku dan mengurutkannya
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # mengumpulkan 10 rekomendasi teratas
    sim_scores = sim_scores[:11]
    book_indices = [i[0] for i in sim_scores]

    # menampilkan 10 hasil rekomendasi buku
    print(f"Recommendation for the book :\n")

    recommend = clean_books.iloc[book_indices]
    print(f"- {book_title} with Genre {recommend['Genre'].iloc[1]} -\n")

    for book in recommend['Title']:
        print(book)
    print('\n')
```

```python
recommended_books = recommend_books('Ravi\'s Roar')
```

```python
recommended_books = recommend_books('Holes')
```
